```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1) #quadratic equation used
plt.scatter(x,y,color='b')
plt.xlabel('X Points')
plt.ylabel('Y Points')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg_Model1=LinearRegression()
reg_Model1.fit(x_train,y_train) #SimpleLinearRegression
```

```python
from sklearn.metrics import r2_score
score1=r2_score(y_test,reg_Model1.predict(x_test))
print(score1)
```

```python
plt.plot(x_train,reg_Model1.predict(x_train),color='r')
plt.scatter(x_train,y_train)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
p=PolynomialFeatures(degree=2,include_bias=True) #Beta 0 value 1, else 0, try for degree 3 and 4 also
x_train_p=p.fit_transform(x_train)
x_test_p=p.transform(x_test) # we don't fit test data
x_train_p #beta 0 (intercept), beta 1 for x and beta 2 for x^2 printed
```

```python
x_test_p
```

```python
reg_Model2=LinearRegression()
reg_Model2.fit(x_train_p,y_train)
y_predict2=reg_Model2.predict(x_test_p)
score2=r2_score(y_test,y_predict2)
score2 #r2 score improved
```

```python
print(reg_Model2.coef_)
print(reg_Model2.intercept_)
```

```python
plt.scatter(x_train,reg_Model2.predict(x_train_p))
plt.scatter(x_train,y_train)
plt.show()
```

```python
def sortswas(ar1, ar2):
  c = 0
  for i in range(len(ar1)):
    c+=1
    for j in range(0, len(ar1) - i - 1):
      if ar1[j] > ar1[j + 1]:
        temp = ar1[j]
        ar1[j] = ar1[j+1]
        ar1[j+1] = temp
        temp = ar2[j]
        ar2[j] = ar2[j+1]
        ar2[j+1] = temp
  # print(c)
```
